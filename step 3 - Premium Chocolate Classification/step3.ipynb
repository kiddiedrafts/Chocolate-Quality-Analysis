```python
import numpy as np
import pandas as pd

df = pd.read_csv('chocolate_price.csv')
df.head()
```

```python
df = df[df['Cocoa Percent'] <= 70]
df.reset_index(drop = True, inplace = True)
df
```

```python
grouped = df.groupby(['Company', 'Review Date'])['Rating'].mean()
companies = grouped.unstack()
companies
```

```python
companies.columns.name = None
companies.index.name = None
```

```python
filtered_df = df[df['Review Date'] >= 2012]
mean_ratings = filtered_df.groupby('Company')['Rating'].mean()
mean_ratings.index.name = None
mean_ratings
```

```python
best_ratings = mean_ratings.sort_values(ascending=False, kind='mergesort')
best_ratings = best_ratings.head(10)
mean_ratings.index.name = None
best_ratings
```

```python
chocolates_to_sell = df[df['Company'].isin(best_ratings.index)]
chocolates_to_sell = chocolates_to_sell.reset_index(drop = True)
chocolates_to_sell
```

```python
priceSum = chocolates_to_sell['price(100g)'].sum()
priceSum
```
